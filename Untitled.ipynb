```python
import numpy as np
import matplotlib.pyplot as plt

class Perceptron:
    def __init__(self, input_size, learning_rate=0.01, max_epochs=1000):
        """
        Initialize a Perceptron model
        
        Parameters:
        input_size (int): Number of input features
        learning_rate (float): Learning rate for weight updates
        max_epochs (int): Maximum number of training epochs
        """
        # Initialize weights (including bias) with small random values
        self.weights = np.random.randn(input_size + 1) * 0.01
        self.learning_rate = learning_rate
        self.max_epochs = max_epochs
        
    def activate(self, x):
        """Step activation function"""
        return 1 if x >= 0 else 0
    
    def predict(self, X):
        """
        Make predictions for input samples
        
        Parameters:
        X (numpy.ndarray): Input samples of shape (n_samples, input_size)
        
        Returns:
        numpy.ndarray: Predicted outputs
        """
        # Add bias term (1) to input
        X_with_bias = np.c_[np.ones(X.shape[0]), X]
        
        # Calculate net input
        net_input = np.dot(X_with_bias, self.weights)
        
        # Apply activation function
        predictions = np.array([self.activate(net) for net in net_input])
        return predictions
    
    def fit(self, X, y):
        """
        Train the perceptron on the given data
        
        Parameters:
        X (numpy.ndarray): Training samples of shape (n_samples, input_size)
        y (numpy.ndarray): Target values
        
        Returns:
        list: Number of misclassifications in each epoch
        """
        errors_per_epoch = []
        
        for epoch in range(self.max_epochs):
            errors = 0
            
            # Add bias term (1) to input
            X_with_bias = np.c_[np.ones(X.shape[0]), X]
            
            for xi, target in zip(X_with_bias, y):
                # Calculate prediction
                net_input = np.dot(xi, self.weights)
                prediction = self.activate(net_input)
                
                # Update weights if prediction is wrong
                error = target - prediction
                if error != 0:
                    # Update all weights
                    self.weights += self.learning_rate * error * xi
                    errors += 1
            
            # Store number of errors
            errors_per_epoch.append(errors)
            
            # Early stopping if no errors (perfect fit)
            if errors == 0:
                print(f"Converged after {epoch+1} epochs")
                break
                
        return errors_per_epoch

# # Example: Implementing AND logic gate
# def demonstrate_perceptron():
#     # AND logic gate training data
#     X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
#     y = np.array([0, 0, 0, 1])  # AND gate output
    
#     # Create and train perceptron
#     perceptron = Perceptron(input_size=2, learning_rate=0.1, max_epochs=100)
#     errors = perceptron.fit(X, y)
    
#     # Print final weights
#     print(f"Final weights (bias, w1, w2): {perceptron.weights}")
    
#     # Test the perceptron
#     print("\nPredictions:")
#     predictions = perceptron.predict(X)
#     for x_i, pred, actual in zip(X, predictions, y):
#         print(f"Input: {x_i}, Predicted: {pred}, Actual: {actual}")
    
#     # Plot decision boundary
#     plot_decision_boundary(perceptron, X, y)
    
#     # Plot errors per epoch
#     plt.figure(figsize=(10, 5))
#     plt.plot(range(1, len(errors) + 1), errors, marker='o')
#     plt.xlabel('Epochs')
#     plt.ylabel('Number of misclassifications')
#     plt.title('Perceptron Learning')
#     plt.grid(True)
#     plt.show()

# def plot_decision_boundary(perceptron, X, y):
#     # Plot the decision boundary
#     plt.figure(figsize=(10, 6))
    
#     # Plot the training points
#     plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='red', marker='o', label='Class 0')
#     plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', marker='x', label='Class 1')
    
#     # Define a grid of points
#     x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
#     y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
#     xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
#                          np.arange(y_min, y_max, 0.02))
    
#     # Compute predictions for all grid points
#     grid = np.c_[xx.ravel(), yy.ravel()]
#     Z = perceptron.predict(grid)
#     Z = Z.reshape(xx.shape)
    
#     # Plot the decision boundary
#     plt.contourf(xx, yy, Z, alpha=0.3)
#     plt.xlim(xx.min(), xx.max())
#     plt.ylim(yy.min(), yy.max())
    
#     # Plot the decision boundary line
#     w1, w2, b = perceptron.weights[1], perceptron.weights[2], perceptron.weights[0]
#     if w2 != 0:
#         slope = -w1 / w2
#         intercept = -b / w2
#         x_vals = np.array([x_min, x_max])
#         y_vals = slope * x_vals + intercept
#         plt.plot(x_vals, y_vals, 'k-', label='Decision Boundary')
    
#     plt.xlabel('Feature 1')
#     plt.ylabel('Feature 2')
#     plt.title('Perceptron Decision Boundary')
#     plt.legend()
#     plt.tight_layout()
#     plt.show()

if __name__ == "__main__":
    demonstrate_perceptron()
```
